```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url='https://comic.naver.com/webtoon/weekdayList?week=mon'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')
a_tags=soup.select('ul.category_tab a')[1:]

weeks = []
week_link=[]

for a_tag in a_tags:
    weeks.append(a_tag.text[:-3]) # [월,화,수,목,금,토,일,매일]
    week_link.append('https://comic.naver.com' + a_tag['href']) # [각 요일별 url주소]

data={'제목':[],'작가':[], '평점':[], '요일':[]}

for week, url_target in zip(weeks, week_link):
    resp=requests.get(url_target)
    soup=BeautifulSoup(resp.content, 'lxml')
    title_tags=soup.select('ul.img_list li dl dt a')
    author_tags=soup.select('ul.img_list dd.desc a')
    rating_tags=soup.select('ul.img_list div.rating_type strong')

    for title, author, rating in zip(title_tags, author_tags, rating_tags):
        data['제목'].append(title['title'])
        data['작가'].append(author.text)
        data['평점'].append(rating.text)
        data['요일'].append(week)

df=pd.DataFrame(data)
df
```

```python
for idx,src in enumerate(img_srcs):
    # 덮어씨워서 지워지는걸 방지하기 위해 index값을 가져와서 index 값으로 저장하기 위해
    resp=requests.get(src)

    with open('thumb_img_{}.jpg'.format(idx),'wb') as f:
            # 'thumb_img_' + str(idx) + '.jpg'
            # 'thumb_img_{}.jpg'.format(idx)
        f.write(resp.content)
    break # 하나만 해보기위해
    # 순서대로 되지가 않음
```

```python
url='https://finance.naver.com/sise/sise_index_day.naver?code=KPI200'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')
```

```python
# 날짜
날짜_list=[]
날짜_tags=soup.select('tr td.date')

for i in 날짜_tags:
    날짜_list.append(i.text.strip())
날짜_list
```

```python
# 체결가 등락률 거래량 거래대금
tab_1=soup.select('tr td.number_1')

체결가=tab_1[::4]
등락률=tab_1[1::4]
거래량=tab_1[2::4]
거래대금=tab_1[3::4]

체결가_list=[]
등락률_list=[]
거래량_list=[]
거래대금_list=[]

for i,j,k,l in zip(체결가, 등락률, 거래량, 거래대금):
    체결가_list.append(i.text.strip())
    등락률_list.append(j.text.strip())
    거래량_list.append(k.text.strip())
    거래대금_list.append(l.text.strip())

print(체결가_list)
print(등락률_list)
print(거래량_list)
print(거래대금_list)
```

```python
# columns 만들기
열이름=[]

for i in soup.select('tr th'):
    열이름.append(i.text.strip())
열이름
```

```python
# 전일비 만들기
전일비_tags=soup.select('td.rate_down span.tah.p11')
rate_down_list=[]
for i in 전일비_tags:
    rate_down_list.append(float(i.text.strip()))


화살표_tags=soup.select('td.rate_down img')
화살표_list=[]
for i in 화살표_tags:
    화살표_list.append((i['src'].split("_")[-1].strip('.gif')))

전일비_list=[]

for i,j in zip(화살표_list, rate_down_list):
    if i=='down':
        전일비_list.append(-j)
    else:
        전일비_list.append(j)


print(rate_down_list)
print(화살표_list)
print(전일비_list)
```

과제: 코스피 200 전체페이지

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
# 마지막 페이지 추출하기
url='https://finance.naver.com/sise/sise_index_day.naver?code=KPI200&page=1'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')
last_pg=int(soup.select('td.pgRR a')[0]['href'].split('page=')[-1])

df_list=[]

for pg in tqdm(range(1,last_pg+1)):
    url='https://finance.naver.com/sise/sise_index_day.naver?code=KPI200&page={}'.format(pg)
    resp=requests.get(url)
    soup=BeautifulSoup(resp.content, 'lxml')

    data={}

    summary_tags=soup.select('tr th') # 날짜, 체결가 전일비, 등락률, 거래량(천주), 거래대금(백만)

    for key in summary_tags: # 항목 이름 만들기(딕셔너리 key추가)
        data[f'{key.text}']=[]

    date_tags=soup.select('td.date') # 날짜 태그들
    price_tags=soup.select('td.number_1')[0::4] # 체결가 태그들
    rate_days=soup.select('td.rate_down span.tah.p11') # 전일비 태그들 인데 span.tah.p11 까지 가져오면 마지막페이지에서 전일비만 5개를 가져옴
    rate_per=soup.select('td.number_1')[1::4] # 등락률 태그들
    trade_vol=soup.select('td.number_1')[2::4] # 거래량 태그들
    trade_pri=soup.select('td.number_1')[3::4] # 거래대금 태그들
        
    tags_len=min([len(date_tags), len(price_tags), len(rate_days), len(rate_per), len(trade_vol), len(trade_pri)])
    # 전일비 태그들 인데 span.tah.p11 까지 가져오면 마지막페이지에서 다른애들은 6개를 가져와도 전일비만 5개를 가져옴

    for i in range(tags_len): # 비어있는 data dictionary 채우기
        data['날짜'].append(date_tags[i].text)
        data['체결가'].append(price_tags[i].text)

        if rate_days[i]['class'][-1]=='red02':
            data['전일비'].append("+"+rate_days[i].text.strip())
        else:
            data['전일비'].append("-"+rate_days[i].text.strip())

        data['등락률'].append(rate_per[i].text.strip())
        data['거래량(천주)'].append(trade_vol[i].text)
        data['거래대금(백만)'].append(trade_pri[i].text)

    df=pd.DataFrame(data)
    df_list.append(df)

df1=pd.concat(df_list, ignore_index=True)
```

과제: 네이버 웹툰 장르별 제목/작가/평점

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

url='https://comic.naver.com/webtoon/genre?genre=episode'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')

data={}
genre_list=[]
genre_kor_list=[]
genre_tags=soup.select('ul.spot li a')

for i in tqdm(genre_tags):

    genre_kor=i.text.strip()
    genre_kor_list.append(genre_kor)

    genre_eng=i['href'].split('genre=')[-1]
    genre_list.append(genre_eng)

    data[f'{genre_eng}']=[]

for genre in tqdm(genre_list):

    url='https://comic.naver.com/webtoon/genre?genre={}'.format(genre)
    resp=requests.get(url)

    soup=BeautifulSoup(resp.content, 'lxml')
    title_tags=soup.select('ul li dl dt a')
    author_tags=soup.select('ul li dl dd.desc a')
    rating_tags=soup.select('div.rating_type strong')

    for i in range(len(title_tags)):
        title=title_tags[i]['title']
        author=author_tags[i].text
        rating=rating_tags[i].text.strip()
        data[f'{genre}'].append(f'작가: {title} \n 제목: {author} \n 평점: {rating}')

data1={key:pd.Series(value) for key, value in data.items()}
# 리스트와 딕셔너리는 서로 길이가 다르면 데이터프레임으로 만들어지지 않지만 시리즈는 길이가 달라도 빈곳이 NaN칸으로 채워져서 만들어진다.

df=pd.DataFrame(data1)
df.columns=genre_kor_list
```

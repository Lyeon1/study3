```python
html=resp.content
# html에 resp의 html을 담음
```

- resp.text : 문자열 형식 -> 사람친화적
- resp.content : 바이너리 형식 -> 기계친화적
- 한글의 경우 text로 하면 디코딩할때 깨질 수도 있음

```python
soup=BeautifulSoup(html, 'lxml')
# BeautifulSoup 형태로 변환, 파싱(Parsing)방법: lxml / lxml은 파싱 프로그램이름으로 가장 효율적이라고 알려져 있음
# html타입을 pyhton에서 다루기 위해 변경해주는 작업

tags=soup.select('a.news_tit')
# find, find_all보다 효율적인 함수. select를 쓰자
# . 은 클래스 " "은 하위태그

# ul class='a b c' -> a의 하위클래스 b의 하위클래스 c
# select(ul.a.b.c) -> 불러오는 방법

# 클래스가 news_area인 div 태그 안에 있는 a 태그

type(tags), len(tags)
# 길이확인을 해서 본래 가져오가자 한것들의 개수와 일치하는지 확인
```

```python
list=['mon','tue','wed','tur','fri','sat','sun']

for day in list:
    
    url='https://comic.naver.com/webtoon/weekdayList?week={}'.format(day)
    resp=requests.get(url)
    soup=BeautifulSoup(resp.content, 'lxml')
    tags_name=soup.select('dl dt a')
    
    name=[]
    for cartoon in tags_name:
        name.append(cartoon.text)
    print("{} : {}".format(day,name))
    print("{} 웹툰 개수 : {}".format(day,len(name)))

# 페이지 상단의 추천웹툰도 가져와짐

#dl dt a 가 제목인데

#하단 추천웹툰에 떠있는거
# <a href="/webtoon/list?titleId=785749&amp;weekday=mon" title="슈퍼스타 천대리">슈퍼스타 천대리</a>

# 상단 전체웹툰에 떠있는거(dl dt a의 하위태그에 속해있음)
# <strong title="슈퍼스타 천대리">슈퍼스타 천대리</strong>
```

```python
# 해결방법
# 상위 태그로 이동하면서 속성값으로 class 나 id 를 가지는 상위 태그까지 경로 넣어주기
list=['mon','tue','wed','tur','fri','sat','sun']

for day in list:
    
    url='https://comic.naver.com/webtoon/weekdayList?week={}'.format(day)
    resp=requests.get(url)
    soup=BeautifulSoup(resp.content, 'lxml')
    tags_name=soup.select('ul.img_list li dl dt a')
    
    name=[]
    for cartoon in tags_name:
        name.append(cartoon.text)
    print("{} : {}".format(day,name))
    print("{} 웹툰 개수 : {}".format(day,len(name)))
```

```python
list=['mon','tue','wed','tur','fri','sat','sun','dailyplus']

data={'day' : [],'name' : [], 'writer' : [], 'point' : []}

for day in list:

    url='https://comic.naver.com/webtoon/weekdayList?week={}'.format(day)
    resp=requests.get(url)
    soup=BeautifulSoup(resp.content, 'lxml')
    tags_name=soup.select('ul.img_list li dl dt a') # 웹툰 제목
    tags_writer=soup.select('ul.img_list li dl dd.desc a') # 작가
    tags_point=soup.select('div.rating_type strong') # 별점

    for i in tags_name:
        data['day'].append(day)
        data['name'].append(i.text)

    for i in tags_writer:
        data['writer'].append(i.text)

    for i in tags_point:
        data['point'].append(i.text)

df=pd.DataFrame(data)
df
```

원하는 네이버웹툰 정해서 모든 화의 베스트 댓글

```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
from tqdm import tqdm
```

```python
# 최신화까지의 회차 구하기
url='https://comic.naver.com/webtoon/list?titleId=778963&weekday=tue&page=1'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')
last_tit=soup.select('td.title a')[0]['href'].split('&no=')[-1].split('&')[0] # 회차태그들에서 가장 처음의 링크에서 스플릿2번해서 숫자가져옴

# 모든회차에 접근하기

webtoon_df_list=[]

for tit in tqdm(range(1,int(last_tit)+1)):
    url='https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?'

    headers={'referer' : 'https://comic.naver.com/comment/comment?titleId=778963&no=1',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}

    parmas={
            'ticket':'comic',
            'templateId':'webtoon',
            'pool':'cbox3',
            '_callback':'',
            'lang':'ko',
            'country':'KR',
            'objectId':'778963_{}'.format(tit),
            'pageSize':15,
            'indexSize':10,
            'listType':'OBJECT',
            'pageType':'default',
            'page':1,
            'sort':'best'
            }

    resp=requests.get(url, headers=headers, params=parmas)
    result=json.loads(resp.text[10:-2])
    result=result['result']['commentList']

    webtoon={'회차':[],
             '댓글':[],
             '좋아요':[],
             '싫어요':[]}

    for comment in result:
        webtoon['회차'].append(tit)
        webtoon['댓글'].append(comment['contents'])
        webtoon['좋아요'].append(comment['sympathyCount'])
        webtoon['싫어요'].append(comment['antipathyCount'])

    webtoon_df_list.append(pd.DataFrame(webtoon))
```

```python
df=pd.concat(webtoon_df_list, ignore_index=True)
df
```

원하는 네이버웹툰 정해서 모든 화의 전체 댓글

```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
from tqdm import tqdm
```

```python
# 최신화까지의 회차 구하기
url='https://comic.naver.com/webtoon/list?titleId=778963&weekday=tue&page=1'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')
last_tit=soup.select('td.title a')[0]['href'].split('&no=')[-1].split('&')[0] # 회차태그들에서 가장 처음의 링크에서 스플릿2번해서 숫자가져옴

# 모든회차에 접근하기

webtoon_df_list=[]

for tit in tqdm(range(1,int(last_tit)+1)):
    url='https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?'

    headers={'referer' : 'https://comic.naver.com/comment/comment?titleId=778963&no=1',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}

    parmas={
            'ticket':'comic',
            'templateId':'webtoon',
            'pool':'cbox3',
            '_callback':'',
            'lang':'ko',
            'country':'KR',
            'objectId':'778963_{}'.format(tit),
            'pageSize':15,
            'indexSize':10,
            'listType':'OBJECT',
            'pageType':'default',
            'page':1,
            'sort':'NEW'
            }

    resp=requests.get(url, headers=headers, params=parmas)
    result=json.loads(resp.text[10:-2])
    total_page=result['result']['pageModel']['totalPages'] # 전체댓글 페이지 수
    
    for pg in range(1,total_page+1):
        url='https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?'
        headers={'referer': 'https://comic.naver.com/comment/comment?titleId=778963&no=1',
                 'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
        params={
                'ticket':'comic',
                'templateId':'webtoon',
                'pool':'cbox3',
                '_callback':'',
                'lang':'ko',
                'country':'KR',
                'objectId':'778963_{}'.format(tit),
                'pageSize':'15',
                'indexSize':'10',
                'listType':'OBJECT',
                'pageType':'default',
                'page':pg,
                'currentPage':'2',
                'refresh':'false',
                'sort':'NEW',
                }

        resp=requests.get(url, headers=headers, params=parmas)
        result=json.loads(resp.text[10:-2])
        result=result['result']['commentList']

        webtoon={'회차':[],
                '댓글':[],
                '좋아요':[],
                '싫어요':[]}

        for comment in result:
            webtoon['회차'].append(tit)
            webtoon['댓글'].append(comment['contents'])
            webtoon['좋아요'].append(comment['sympathyCount'])
            webtoon['싫어요'].append(comment['antipathyCount'])

        webtoon_df_list.append(pd.DataFrame(webtoon))
```

전체 모든 웹툰, 최신화, 베스트 댓글

```python
import requests
import json
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
# 요일 리스트 가져오기
url='https://comic.naver.com/webtoon/weekdayList?week=mon'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')
week_tags=soup.select('ul.category_tab li a')[1:]

weeks=[]
for week_tag in week_tags:
    weeks.append(week_tag['href'].split('week=')[1])

webtoon_df_list=[]

for week in tqdm(weeks):

    url='https://comic.naver.com/webtoon/weekdayList?week={}'.format(week)
    resp=requests.get(url)
    soup=BeautifulSoup(resp.content, 'lxml')
    title_tags=soup.select('ul.img_list dl dt a')
    
    titleID_list=[]
    title_name_list=[]
    for titleID in title_tags: # 웹툰 titleID
        titleID_list.append(titleID['href'].split('titleId=')[1].split('&')[0])
        title_name_list.append(titleID['title'])

    title_list=zip(title_name_list,titleID_list)

    for title, titleID in tqdm(title_list):
        url='https://comic.naver.com/webtoon/list?titleId={}&weekday={}'.format(titleID, week)
        resp=requests.get(url)
        soup=BeautifulSoup(resp.content, 'lxml')
        last_tit=soup.select('td.title a')[0]['href'].split('no=')[1].split('&')[0] # 최신화

        url='https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?'

        headers={'referer': 'https://comic.naver.com/comment/comment?titleId={}&no={}'.format(titleID,last_tit),
                 'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36'}
        params={
                'ticket':'comic',
                'templateId':'webtoon',
                'pool':'cbox3',
                '_callback':'',
                'lang':'ko',
                'country':'KR',
                'objectId':'{}_{}'.format(titleID,last_tit),
                'pageSize':15,
                'indexSize':10,
                'listType':'OBJECT',
                'pageType':'default',
                'page':1,
                'refresh':'true',
                'sort':'best'
                }
        resp=requests.get(url, headers=headers, params=params)
        result=json.loads(resp.content[10:-2])
        result=result['result']['commentList']


        webtoon={'요일':[],
                 '제목':[],
                '최신회차':[],
                '댓글':[],
                '좋아요':[],
                '싫어요':[]}
        

        for comment in result:
            webtoon['요일'].append(week)
            webtoon['제목'].append(title)
            webtoon['최신회차'].append(last_tit)
            webtoon['댓글'].append(comment['contents']) # 댓글
            webtoon['좋아요'].append(comment['sympathyCount']) # 좋아요
            webtoon['싫어요'].append(comment['antipathyCount']) # 싫어요
        webtoon_df_list.append(pd.DataFrame(webtoon))
```

```python
pd.concat(webtoon_df_list, ignore_index=True)
```

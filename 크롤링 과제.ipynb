```python
url='https://comic.naver.com/webtoon/genre?genre=episode'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')
genre_tags=soup.select('ul.spot a')

genre_dic={}
genre_kor_list=[]
for genre in tqdm(genre_tags):
    genre_eng=genre['href'].split('genre=')[-1]
    genre_dic[f'{genre_eng}']=[]
    genre_kor_list.append(genre.text)

    url='https://comic.naver.com/webtoon/genre?genre={}'.format(genre_eng)
    resp=requests.get(url)
    soup=BeautifulSoup(resp.content, 'lxml')

    title_tags=soup.select('div.thumb a') # 제목
    author_tags=soup.select('dd.desc a') # 작가
    rating_tags=soup.select('div.rating_type strong') # 평점

    for i in range(len(title_tags)):
        title=title_tags[i]['title']
        author=author_tags[i].text
        rating=rating_tags[i].text
        genre_dic[f'{genre_eng}'].append("/".join([title,author,rating]))

data={key:pd.Series(value) for key, value in genre_dic.items()}
    
```

```python
df=pd.DataFrame(data)
df.columns=genre_kor_list
df
```

```python
url='https://finance.naver.com/sise/sise_index_day.naver?code=KPI200&page=1'
resp=requests.get(url)
soup=BeautifulSoup(resp.content, 'lxml')

columns_tags=soup.select('table.type_1 tr th') # 항목

# 맨 뒤 페이지 가져오기
last_pg=soup.select('td.pgRR a')[0]['href'].split('page=')[-1]

df_list=[]

for pg in tqdm(range(1,int(last_pg)+1)):
    url='https://finance.naver.com/sise/sise_index_day.naver?code=KPI200&page={}'.format(pg)
    resp=requests.get(url)
    soup=BeautifulSoup(resp.content, 'lxml')

    dic={} # 
    for i in columns_tags:
        dic[f'{i.text}']=[]

    date_tags=soup.select('td.date') # 날짜
    price_tags=soup.select('td.number_1')[::4] # 체결가
    rate_per_tags=soup.select('td.number_1')[1::4] # 등락률 strip()필요
    vol_tags=soup.select('td.number_1')[2::4] # 거래량
    price_vol_tags=soup.select('td.number_1')[3::4] # 거래대금
    rate_day_tags=soup.select('td.rate_down span.tah.p11') # 전일비

    for_range=min(len(date_tags), len(price_tags), len(rate_per_tags), len(vol_tags), len(price_vol_tags), len(rate_day_tags))

    for i in range(for_range):
        if rate_day_tags[i]['class'][-1]=='nv01':
            dic['전일비'].append("-"+rate_day_tags[i].text.strip())
        else:
            dic['전일비'].append(rate_day_tags[i].text.strip())
        dic['날짜'].append(date_tags[i].text)
        dic['체결가'].append(price_tags[i].text)
        dic['등락률'].append(rate_per_tags[i].text.strip())
        dic['거래량(천주)'].append(vol_tags[i].text)
        dic['거래대금(백만)'].append(rate_day_tags[i].text.strip())

    df=pd.DataFrame(dic)
    df_list.append(df)
```

```python
total_df=pd.concat(df_list, ignore_index=True)
total_df
```

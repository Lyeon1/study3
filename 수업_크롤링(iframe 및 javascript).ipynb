```python
import requests
url = 'https://www.bluer.co.kr/api/v1/restaurants?page=0&size=30&query=&foodType=&foodTypeDetail=&feature=&location=&locationDetail=&area=&areaDetail=&priceRange=&ribbonType=&recommended=false&isSearchName=false&tabMode=single&searchMode=ribbonType&zone1=&zone2=&zone2Lat=&zone2Lng='
resp = requests.get(url)
import json
result=json.loads(resp.content)

restaurants=result['_embedded']['restaurants']

r_list=[]

import pandas as pd
from tqdm import tqdm

for restaurant in tqdm(restaurants):
    blue_info = {
    '가게명' : restaurant['headerInfo']['nameKR'],
    '리본등급' : restaurant['headerInfo']['ribbonType'],
    "사이트" : restaurant['defaultInfo']['website'], 
    "전화번호": restaurant['defaultInfo']['phone'], 
    "휴무": restaurant['defaultInfo']['dayOff'],
    "주차":restaurant['statusInfo']['parking'],
    "가격범위":restaurant['statusInfo']['priceRange'],
    "메뉴":restaurant['statusInfo']['menu'],
    "영업시간":restaurant['statusInfo']['businessHours']
    }
    r_list.append(blue_info)
    
import pandas as pd
df=pd.DataFrame(r_list)
df.info()
```

위의 코드는 리스트 내부값 1개가 가게 1개의 모든 정보를 담고있고

아래의 코드는 이름리스트에 모든 가게이름, 리본리스트에 모든 리본데이터가 들어감

같은 데이터프레임 형태로 나오는건 pandas의 자체적인 기능인듯?

```python
import requests
url = 'https://www.bluer.co.kr/api/v1/restaurants?page=0&size=30&query=&foodType=&foodTypeDetail=&feature=&location=&locationDetail=&area=&areaDetail=&priceRange=&ribbonType=&recommended=false&isSearchName=false&tabMode=single&searchMode=ribbonType&zone1=&zone2=&zone2Lat=&zone2Lng='
resp = requests.get(url)
import json
result=json.loads(resp.content)

restaurants=result['_embedded']['restaurants']

names, ribbons, sites, tels, dayoffs, parks, prices, menus, times = [], [], [], [], [], [], [], [], []

import pandas as pd
from tqdm import tqdm


for restaurant in tqdm(restaurants):
    names.append(restaurant["headerInfo"]["nameKR"])
    ribbons.append(restaurant["headerInfo"]["ribbonType"])
    sites.append(restaurant["defaultInfo"]["website"])
    tels.append(restaurant["defaultInfo"]["phone"])
    dayoffs.append(restaurant["defaultInfo"]["dayOff"])
    parks.append(restaurant['statusInfo']["parking"])
    prices.append(restaurant['statusInfo']["priceRange"])
    menus.append(restaurant['statusInfo']["menu"])
    times.append(restaurant['statusInfo']["businessHours"])

    blue_info = {
        '가게명' : names,
        '리본등급' : ribbons,
        "사이트" : sites, 
        "전화번호": tels, 
        "휴무": dayoffs,
        "주차": parks,
        "가격범위": prices,
        "메뉴": menus,
        "영업시간": times
    }

blue_info
df=pd.DataFrame(blue_info)
df.info()
```

```python
# 하나의 화 전체 댓글 가져오기
import requests
import json
from tqdm import tqdm

url='https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?'

headers={
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36',
    'referer' : 'https://comic.naver.com/comment/comment?titleId=783053&no=52'
}

params={ # url의 ? 뒤의 쿼리스트링을
        'ticket':'comic',
        'templateId':'webtoon',
        'pool':'cbox3',
        # _cv=20221018144728 얘는 생략하네?
        '_callback':'', #jQuery1124008606681137875127_1667201134255 이였으나 requests로 읽어올때 앞에 쓸데없는거 날리기 위해 null값
        'lang':'ko',
        'country':'KR',
        'objectId':'783053_52',
        # categoryId= # 없는 값들은 버림
        'pageSize':'15',
        'indexSize':'10',
        # groupId=
        'listType':'OBJECT',
        'pageType':'default',
        'page':'1',
        # currentPage=2 # 불필요해서 버림
        'refresh':'True',
        'sort':'NEW' # 전체댓글 URL의 고유 쿼리
    }

resp=requests.get(url, headers=headers, params=params)
# params는 쿼리스트링을 받아주는 부분
result=json.loads(resp.text[10:-2]) # 앞의 _callback( 이랑 뒤의 ); 날림

# # 맨뒤버튼 가져오기
last_pg=result['result']['pageModel']['totalPages'] # -> 93

for pg in tqdm(range(1,last_pg+1)):

    url='https://apis.naver.com/commentBox/cbox/web_naver_list_jsonp.json?'

    headers={
        'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36',
        'referer' : 'https://comic.naver.com/comment/comment?titleId=783053&no=52'
    }

    params={
            'ticket':'comic',
            'templateId':'webtoon',
            'pool':'cbox3',
            '_callback':'',
            'lang':'ko',
            'country':'KR',
            'objectId':'783053_52',
            'pageSize':'15',
            'indexSize':'10',
            'listType':'OBJECT',
            'pageType':'default',
            'page': pg,
            'refresh':'false',
            'sort':'NEW'
        }
    resp=requests.get(url, headers=headers, params=params)
    result=json.loads(resp.text[10:-2])
    commentList=result['result']['commentList']

    comments=[]

    for comment in commentList:
        comments.append(comment['contents'])
    
    print("{}페이지 댓글: {}".format(pg,comments))
```
